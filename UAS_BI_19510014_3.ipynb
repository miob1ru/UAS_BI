```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_excel('/content/caesarian.xlsx')
df
```

```python
import math
dis = []
for i in range(80):
    dis.append(math.sqrt((float(df.iloc[i]['Usia'])-25)**4+
                         (float(df.iloc[i]['Kelahiran_ke-'])- 1)**4+
                         (float(df.iloc[i]['Waktu_Kelahiran'])-0)**4+
                         (float(df.iloc[i]['Tekanan_darah'])-1)**4))
```

```python
df['dis'] = dis
df
```

```python
df.sort_values('dis')
```

```python
y = df.sort_values('dis').head(5)
y
```

```python
z = y["Caesarian"]
z
```

```python
np.mean(z)
```

```python
df.to_excel('/content/output_no2.xlsx')
```
