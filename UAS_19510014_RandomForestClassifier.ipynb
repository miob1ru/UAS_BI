```python
# importing libraries  
import numpy as np  
import matplotlib.pyplot as mtp  
import pandas as pd
```

```python
#importing datasets  
df=pd.read_excel('/content/MOKAS.xlsx')  
df
```

```python
#MENENTUKAN NILAI X_USIA
df['X_USIA']=np.where(df['USIA'] > 30, 3, 0)
df['X_USIA']=np.where((df['USIA'] >= 20) & (df['USIA'] <= 30), 2, df['X_USIA'])
df['X_USIA']=np.where(df['USIA'] < 20, 1, df['X_USIA'])
df
```

```python
#MENENTUKAN NILAI X_JKELAMIN
df['X_JKELAMIN']=np.where(df['JENIS KELAMIN']=='WANITA', 0, 1)
df
```

```python
#MENENTUKAN NILAI X_PEKERJAAN
df['X_PEKERJAAN']=np.where(df['PEKERJAAN']=='PEGAWAI', 1, 0)
df['X_PEKERJAAN']=np.where(df['PEKERJAAN']=='PELAJAR', 2, df['X_PEKERJAAN'])
df['X_PEKERJAAN']=np.where(df['PEKERJAAN']=='WIRASWASTA', 3, df['X_PEKERJAAN'])
df
```

```python
#MENENTUKAN NILAI X_TABUNGAN
df['X_TABUNGAN']=np.where(df['TABUNGAN'] < 20000000, 1, 0)
df['X_TABUNGAN']=np.where((df['TABUNGAN'] >= 20000000) & (df['TABUNGAN'] <= 30000000), 2, df['X_TABUNGAN'])
df['X_TABUNGAN']=np.where(df['TABUNGAN'] > 30000000, 3, df['X_TABUNGAN'])
df
```

```python
#Extracting Independent and dependent Variable  
x=df.iloc[:, [3, 5, 7, 9]].values  
y=df.iloc[:, 10].values  
x
```

```python
# Splitting the dataset into training and test set.  
from sklearn.model_selection import train_test_split  
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size= 0.70, random_state=0)  
x_train
```

```python
#Fitting Decision Tree classifier to the training set  
from sklearn.ensemble import RandomForestClassifier  
model=RandomForestClassifier(n_estimators=30, criterion="entropy")  
model.fit(x_train, y_train)
```

```python
y_pred=model.predict(x_test)  
y_pred
```

```python
#Creating the Confusion matrix  
from sklearn.metrics import confusion_matrix  
cm=confusion_matrix(y_test, y_pred) 
cm
```
